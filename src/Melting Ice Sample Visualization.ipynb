```python
from matplotlib import pyplot as plt, patches
plt.rcParams["figure.figsize"] = [5, 5]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot()
circle1 = patches.Circle((30, 30), radius=20, color='lightgrey')
circle2 = patches.Circle((30, 30), radius=16, color='lightblue')
circle3 = patches.Circle((30, 30), radius=14, color='grey')
ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)
ax.axis('equal')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [5, 5]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot()
```

```python
def sample_func(t):
    return 14+t

duration = 6

def make_frame(t):
    ax.clear()
    circle1 = patches.Circle((30, 30), radius=20, color='lightgrey')
    circle2 = patches.Circle((30, 30), radius=sample_func(t), color='lightblue')
    circle3 = patches.Circle((30, 30), radius=14, color='grey')
    ax.add_patch(circle1)
    ax.add_patch(circle2)
    ax.add_patch(circle3)
    ax.axis('equal')
    return mplfig_to_npimage(fig)
```

```python
from moviepy.editor import VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
```

```python
animation = VideoClip(make_frame, duration = duration)
```

```python
animation.ipython_display(fps = 20, loop = True, autoplay = True)
```
